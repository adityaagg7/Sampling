```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
df.head()
```

```python
from sklearn.preprocessing import normalize
Amount = normalize([df['Amount']])[0]
df['Amount'] = Amount
df = df.iloc[:, 1:]
df.head()
```

```python
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

X = df.drop('Class', axis=1)
y = df['Class']


oversampler = RandomOverSampler()



x_over, y_over = oversampler.fit_resample(X, y)


print(y_over.value_counts())
```

```python
r_df = pd.concat([x_over, y_over], axis=1)
r_df
```

```python
n = int((1.96*1.96 * 0.5*0.5)/(0.05**2))
sampled_df = r_df.sample(n=n, random_state=42)
print(sampled_df)
```

```python
# systematic sampling

interval = 2
systematic_df = r_df.iloc[::interval]
systematic_df
```

```python


from sklearn.model_selection import train_test_split

n = int((1.96*1.96 * 0.5*0.5)/((0.05)**2))

strata = r_df.groupby('Class')


stratified_df = strata.apply(lambda x: x.sample(n))

stratified_df
```

```python

import numpy as np

from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=10, random_state=42).fit(r_df)
cluster_assignments = kmeans.labels_


selected_clusters = [0, 2, 4, 5, 8]

cluster_series = pd.Series(cluster_assignments)


df_cluster_sample = r_df[cluster_series.isin(selected_clusters)]



(df_cluster_sample)
```

```python
# convenience sampling
convenience_sample = pd.concat([r_df.head(380), r_df.tail(380)])
convenience_sample
```
